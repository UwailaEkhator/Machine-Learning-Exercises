The gini index formula is $ 1 - \sum_{i=0}^{c-1} p_{i}(t) ^2$

```python
def gini_index(y):
    # Count the occurrences of each class label
    values, counts = np.unique(y, return_counts=True)
    # Calculate the probability of each class label
    class_probs = counts / len(y)
    # Calculate the Gini index
    gini = 1 - np.sum(np.square(class_probs))
    return gini
```

The information gain formular is $\Delta = I(parent) - I(children)$ <br>
where $I(parent) =  1 - \sum_{i=0}^{c-1} p_{i}(t) ^2$ <br>
and $I(children) = \sum_{j=1}^{k} \frac{N(V_{j})}{N} I(v_{j})$<br>
$I(children)$ in this case can be written as $(\frac{N_{left}}{N_{total}} * Gini(left)) + (\frac{N_{right}}{N_{total}} * Gini(right))$

```python
def information_gain_gini(y, left_indices, right_indices):
    # Calculate the Gini index before the split (I(parent))
    gini_parent = gini_index(y)
    
    # Calculate the Gini index after the split (I(children))
    gini_left = gini_index(y[left_indices])
    gini_right = gini_index(y[right_indices])
    
    # Calculate the weighted average Gini index after the split
    weighted_gini_children = (len(left_indices) / len(y)) * gini_left + (len(right_indices) / len(y)) * gini_right
    
    # Calculate the information gain using Gini index (Delta = I(parent) - I(children))
    gain = gini_parent - weighted_gini_children
    return gain
```

The entropy formula is $ - \sum_{i=0}^{c-1} p_{i}(t)log_{2}p_{i}t$

```python
def entropy(y):
    # Count the occurrences of each class label
    values, counts = np.unique(y, return_counts=True)
    # Calculate the probability of each class label
    class_probs = counts / len(y)
    # Calculate the entropy
    entropy = -np.sum(class_probs * np.log2(class_probs + 1e-10))  # Adding a small value to prevent log(0)
    return entropy
```

Gain ratio = $\frac{Information gain}{Split Info}$<br>
Where split info = $-\sum_{i=1}^{k} \frac{N(V_{i})}{N} log_{2} \frac{N(V_{i})}{N} $ <br> 
which can be written as=
$-(\frac{N_{left}}{N_{total}} log_{2} \frac{N_{left}}{N_{total}} ) + (\frac{N_{right}}{N_{total}} log_{2} \frac{N_{right}}{N_{total}} )$

```python
def gain_ratio(y, left_indices, right_indices):
    # Calculate the information gain
    gain = information_gain_gini(y, left_indices, right_indices)
    # Calculate the split information
    split_info = -((len(left_indices) / len(y)) * np.log2(len(left_indices) / len(y) + 1e-10) +
                   (len(right_indices) / len(y)) * np.log2(len(right_indices) / len(y) + 1e-10))
    # Calculate the gain ratio
    gain_ratio = gain / split_info if split_info != 0 else 0  # Avoid division by zero
    return gain_ratio
```

```python
def scoref(X, y, f, criterion):
    if criterion == 'gini':
        # Calculate the Gini index for the split
        gini_left = gini_index(y[X[:, f] == 0])
        gini_right = gini_index(y[X[:, f] == 1])
        weighted_gini = (len(X[X[:, f] == 0]) / len(y)) * gini_left + (len(X[X[:, f] == 1]) / len(y)) * gini_right
        score = weighted_gini
    elif criterion == 'information_gain':
        # Calculate the information gain for the split
        left_indices = np.where(X[:, f] == 0)[0]
        right_indices = np.where(X[:, f] == 1)[0]
        score = information_gain_gini(y, left_indices, right_indices)
    elif criterion == 'entropy_gain_ratio':
        # Calculate the gain ratio for the split
        left_indices = np.where(X[:, f] == 0)[0]
        right_indices = np.where(X[:, f] == 1)[0]
        score = gain_ratio(y, left_indices, right_indices)
    else:
        raise ValueError("Invalid criterion. Supported values are 'gini', 'information_gain', and 'entropy_gain_ratio'.")
    return score
```

```python
def isSplittable(y):
    #checks if y has more than 1 unique element, if it has just 1 unique element there is no point splitting further
    return len(set([a for a in y]))!=1

def bestClass(y):#this counts and returns the class label that occurs most frequently in the input list 'y'
    h=set([a for a in y])
    best=-1 #stores count of most frequent class
    valueBest=-1 #stores the most frequent class label
    for f in h:
        count=np.sum(y==f)#counts the number of occurrences in current class label 'f' in the input list 'y'
        if count>=best:
            best=count
            valueBest=f
    return valueBest



def createTree(X, y, criterion='gini'):  # Add a default criterion parameter
    if X.shape[0] > 1 and isSplittable(y):
        best = -19000000
        bestindex = -1
        for i in range(X.shape[1]):
            if len(set([h for h in X[:, i]])) > 1:
                if criterion == 'gini':
                    score = scoref(X, y, i, criterion)  # Use scoref with Gini index
                elif criterion == 'information_gain':
                    left_indices = np.where(X[:, i] == 0)[0]
                    right_indices = np.where(X[:, i] == 1)[0]
                    score = information_gain_gini(y, left_indices, right_indices)  # Calculate information gain using Gini index
                elif criterion == 'entropy_gain_ratio':
                    left_indices = np.where(X[:, i] == 0)[0]
                    right_indices = np.where(X[:, i] == 1)[0]
                    score = gain_ratio(y, left_indices, right_indices)  # Calculate gain ratio using Gini index
                else:
                    raise ValueError("Invalid criterion. Supported values are 'gini', 'information_gain', and 'entropy_gain_ratio'.")
                
                if best <= score:
                    best = score
                    bestindex = i

        if bestindex == -1:
            return ('l', bestClass(y))
        else:
            cr = createTree(X[X[:, bestindex] == 0], y[X[:, bestindex] == 0], criterion)  # Recur with the selected criterion
            cl = createTree(X[X[:, bestindex] == 1], y[X[:, bestindex] == 1], criterion)  # Recur with the selected criterion
            return ('n', (bestindex, cr, cl))
    else:
        return ('l', bestClass(y))
```
